# 3.5 Outlier Exploration

 We decided to not remove any outlier entries. The justifications are:-

 No Outliers in Likert Data: Likert scales represent ordinal data within a fixed range (e.g., 0-4). Each value corresponds to a specific qualitative judgment (e.g., strongly disagree to strongly agree). Outliers are extreme values that deviate significantly from the data distribution, but for Likert data, values are predefined and bounded by the scale. It is impossible to have a value outside this range unless there is an error in data collection or entry. Thus, concepts of outliers in a numerical sense do not apply here, as any value within 0-4 is valid and meaningful.

Binary Target Variable: Binary variables (0 or 1) are categorical and do not have a numerical continuum. This means outliers do not exist for binary data, as all possible values are already defined and intentional. In the context of prediction models, 0 and 1 represent distinct classes, not numerical values that could be unusually high or low.

Practical Implications: Removing outliers is typically aimed at addressing data errors or extreme cases that distort statistical analysis. For Likert and binary data, since values are inherently bounded and meaningful, removing values would lead to data loss rather than correction.

# 4.2 Machine Algorithm Model Fitting

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import VotingClassifier
```

```python
#Split the data into train data and test data
X =df.drop('Divorce_Y_N', axis=1)
y = df['Divorce_Y_N']


scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

```

Define parameters for every algorithm.

```python
# Extra Trees Classifier parameter grid
et_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

# MLP Classifier parameter grid
mlp_param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (100, 50)],
    'activation': ['relu', 'tanh'],
    'solver': ['adam', 'sgd'],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate': ['constant', 'adaptive'],
}

# Bagging Classifier parameter grid
bagging_param_grid = {
    'n_estimators': [10, 20, 30, 50, 100],
    'max_samples': [0.5, 0.7, 1.0],
    'max_features': [0.5, 0.7, 1.0],
    'bootstrap': [True, False],
    'bootstrap_features': [True, False],
    'estimator__max_depth': [None, 5, 10, 20],
}

# SGD classifier parameter grid
sgd_param_grid = {
    'loss': ['perceptron'],
    'penalty': ['elasticnet'],
    'alpha': [0.01, 0.1, 1.0],
    'l1_ratio': [0.25, 0.5, 0.75],
    'max_iter': [1000, 5000],
    'tol': [1e-3, 1e-4, 1e-5],
    'learning_rate': ['optimal', 'constant', 'invscaling', 'adaptive'],
    'eta0': [0.01, 0.1, 1.0],
    'early_stopping': [True, False],
}

# Gradient Boosting Classifier parameter grid
gb_param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'subsample': [0.8, 1.0],
    'max_features': ['auto', 'sqrt', 'log2']
}
# Voting Classifier parameter grid
voting_param_grid = {
    'weights': [
        [1, 1, 1],  # Equal weight for all estimators
        [2, 1, 1],  # Higher weight for the first estimator
        [1, 2, 1],  # Higher weight for the second estimator
        [1, 1, 2]   # Higher weight for the third estimator
    ],
    'voting': ['hard', 'soft'],  # Hard for majority, Soft for probabilities
}
```

Grid search for best hyperparameters for every algorithm

```python

# Extra Trees Classifier Grid Search

et_grid_search = GridSearchCV(ExtraTreesClassifier(), et_param_grid, cv=5)


# MLP Classifier Grid Search

mlp_grid_search = GridSearchCV(MLPClassifier(), mlp_param_grid, cv=5)

#Bagging Classifier Grid Search
bagging_classifier = BaggingClassifier(estimator=DecisionTreeClassifier())
bagging_grid_search = GridSearchCV(bagging_classifier,bagging_param_grid, scoring='accuracy', cv=5)

# SGD Classifier Grid Search
sgd_model = SGDClassifier()
sgd_grid_search = GridSearchCV(estimator=sgd_model, param_grid=sgd_param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)

# Gradient Boosting Classifier Grid Search
gb_grid_search = GridSearchCV(GradientBoostingClassifier(random_state=42), gb_param_grid, cv=5, n_jobs=-1,scoring='accuracy')



```

```python

# Fit the models and find the best hyperparameters

#Extra Tree Classifier
et_grid_search.fit(X_train, y_train)

#Multi Layer Perceptron
mlp_grid_search.fit(X_train, y_train)

#Bagging Classifier
bagging_grid_search.fit(X_train,y_train)

# SGD Classifier
sgd_grid_search.fit(X_train,y_train)

# Gradient Boosting Classifier
gb_grid_search.fit(X_train, y_train)



```

Due to the fact that Voting Classifer requires the best estimators from  Extra Trees best model,MLP best model and Gradient Boosting best model, thus the Voting Classifier will be executed after the three models have fitted.

```python
# Voting Classifier Grid Search
best_et_model = et_grid_search.best_estimator_
best_mlp_model = mlp_grid_search.best_estimator_
best_bagging_model = bagging_grid_search.best_estimator_
best_sgd_model = sgd_grid_search.best_estimator_
best_gb_model = gb_grid_search.best_estimator_
```

```python
# Define the VotingClassifier with the best models of your classifiers
estimators = [
    ('et', best_et_model),  # Extra Trees best model
    ('mlp', best_mlp_model),  # MLP best model
    ('gb', best_gb_model)  # Gradient Boosting best model
]

voting_clf = VotingClassifier(estimators=estimators)

voting_grid_search = GridSearchCV(estimator=voting_clf, param_grid=voting_param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
```

```python
voting_grid_search.fit(X_train, y_train)
```

```python
# Get the best hyperparameters and their respective scores

best_params_extra_trees = et_grid_search.best_params_
best_params_mlp = mlp_grid_search.best_params_
best_params_bagging = bagging_grid_search.best_params_
best_params_sgd = sgd_grid_search.best_params_
best_params_gb = gb_grid_search.best_params_
best_voting_gb = voting_grid_search.best_params_
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, roc_auc_score, roc_curve, confusion_matrix
#Calculate each metric respectively for visualization purpose
def calculate_metrics(y_true, y_pred):
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred)

    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    specificity = tn / (tn + fp)

    roc_auc = roc_auc_score(y_true, y_pred)

    return precision, recall, f1, accuracy, specificity, roc_auc

# Calculate metrics for each model
et_metrics = calculate_metrics(y_test, et_y_pred)
mlp_metrics = calculate_metrics(y_test, mlp_y_pred)
bagging_metrics = calculate_metrics(y_test, bagging_y_pred)
sgd_metrics = calculate_metrics(y_test, sgd_y_pred)
gb_metrics = calculate_metrics(y_test, gb_y_pred)
voting_metrics = calculate_metrics(y_test, voting_y_pred)

# Store metrics in a DataFrame
metrics_df = pd.DataFrame([et_metrics, mlp_metrics, bagging_metrics, sgd_metrics, gb_metrics, voting_metrics],
                         columns=['Precision', 'Recall', 'F1-score', 'Accuracy', 'Specificity', 'ROC AUC'],
                         index=['Extra Trees', 'MLP Classifier', 'Bagging Classifier', 'SGD Classifier', 'Gradient Boosting', 'Voting Classifier'])
```

**Extra Trees Classifier**

Extra Trees Classifier (ETC) has shown promising results in binary classification tasks across various domains. For cardiovascular disease prediction, ETC achieved 90% accuracy, outperforming other methods like Logistic Regression and Support Vector Machine (Rahman et al., 2021). In breast cancer classification, an optimized ETC model attained 99.27% accuracy on the test set, identifying key features such as Size Uniformity and Shape Uniformity (Mathew, 2022). ETC also excelled in stock market prediction, achieving 86.1% accuracy in forecasting significant price changes over a 10-day period (Pagliaro, 2023). A comparative study of tree-based ensemble methods found ETC to be the best classifier for both binary and multi-class problems, with the number of estimators positively correlated to dataset size and complexity (Nurheri Cahyana et al., 2021). These findings highlight ETC's effectiveness and versatility in various binary classification tasks.

**Multi-layer perceptron Classifier**

Multi-layer perceptron (MLP) neural networks have shown promising results in binary classification tasks. For heart disease classification, MLP outperformed other neural network models and statistical techniques, achieving high accuracy, sensitivity, and specificity (Raut & Dudul, 2009). To address class imbalance issues, a cost-sensitive MLP (CMMLP) approach was proposed, which modifies the loss function to prioritize minority class accuracy, demonstrating improved performance on imbalanced datasets (Liu et al., 2018). An efficient training algorithm for MLPs with binary output was developed using singular value decomposition and conjugate gradient methods (Fanelli et al., 1993). MLPs can also be applied to multi-label classification tasks, such as predicting gene function. A single multi-output MLP showed different classification characteristics compared to multiple binary classifiers, including a more rapid decrease in sensitivity as the output cutoff value increased (Skabar et al., 2006). These studies highlight the versatility and effectiveness of MLPs in various binary and multi-label classification scenarios.

**Bagging Classifier**

Bagging classifiers have gained popularity for binary classification, particularly in handling imbalanced datasets (Singhal et al., 2018). These ensemble methods improve classification performance by aggregating multiple models to generalize predictions (Singhal et al., 2018; Parande & Banga, 2020). In diabetes classification, bagged logistic regression has shown superior performance compared to traditional techniques (Parande & Banga, 2020). Binary classification is widely used in various applications, including sockpuppet detection and text categorization (Kumari & Srivastava, 2017). When dealing with imbalanced data, ensemble-based approaches like bagging and boosting have been extensively studied (Kaur & Gosain, 2019). A comparative analysis of various ensemble methods on imbalanced datasets revealed that RusBoost outperformed other ensembles in terms of Area Under the Curve (AUC) performance (Kaur & Gosain, 2019). Overall, bagging classifiers offer a robust solution for improving accuracy and addressing class imbalance in binary classification problems.

**Stochastic Gradient Descent Classifier**

Stochastic Gradient Descent (SGD) has shown promising results in binary classification tasks across various domains. In breast cancer diagnosis, a hybrid approach combining SGD with self-organizing maps outperformed standalone SGD and other classifiers like SVM and Random Forests (Mittal et al., 2015). For Bangla text categorization, SGD demonstrated higher accuracy compared to SVM and Naive Bayes classifiers (Kabir et al., 2015). In dengue classification, binary SGD achieved 86.49% accuracy, while multi-class SGD reached 91.63% accuracy (D. G. & D. A., 2023). To address SGD's termination issues in binary classification, a new computationally inexpensive criterion was proposed for constant step-size SGD with logistic and hinge loss functions. This method showed effectiveness for Gaussian-distributed data and exhibited good predictability on accuracy and running time for both synthetic and real datasets (Baghal et al., 2020).

**Gradient Boosting Classifier**

The family of boosting methods is based on a constructive strategy of ensemble formation. The main idea of boosting is to add new models to the ensemble sequentially. At each iteration, a new weak, base-learner model is trained with respect to the error of the whole ensemble learnt so far (Schapire, 2002). Gradient-descent based formulations were later derived (Freund and Schapire, 1997; Friedman et al., 2000; Friedman, 2001), leading to Gradient Boosting Machines (GBMs). GBMs consecutively fit new models to improve the accuracy of the response variable estimate, using base-learners correlated with the negative gradient of a loss function. The choice of loss functions introduces flexibility and customization for data-driven tasks, and GBMs have achieved notable success in practical applications and data-mining challenges (Bissacco et al., 2007; Hutchinson et al., 2011; Pittman and Brown, 2011; Johnson and Zhang, 2012).

The gradient boosting algorithm (BOOST) is an alternative in genome-assisted evaluations involving large datasets of genotypes and phenotypes. It performs variable selection, uses additive regression models, and is computationally efficient. Initially introduced as AdaBoost by Freund and Schapire (1996), BOOST has been applied to genome-wide prediction of genetic merit and disease susceptibility in animal breeding, demonstrating accuracy comparable to or better than methods like Bayes A or Bayesian LASSO (González-Recio et al., 2010; González-Recio and Forni, 2011). It is well-suited for analyzing high-dimensional and fuzzy data, such as high-density SNP chips, without assuming specific inheritance models.

Gradient boosting algorithms can incorporate several types of boosting techniques, each with unique characteristics tailored for different applications. The several types are XGBoost, Light Gradient Boosting Machine, CatBoost. XGBoost is a machine learning algorithm based on decision-trees that uses a gradient boosting framework. It was a research project which was undertaken at the University of Washington (Chen & Guestrin, 2016). The two main standouts for XGBoost are System Optimizations and Algorithmic Enhancements (Chen & Guestrin, 2016). Light Gradient Boosting Machine is a robust gradient boosting framework that is used for classification,ranking and many other machine learning tasks. It is a distributed, high-performance gradient boosting framework based on decision trees. The algorithm attempted to be a competitor to XGBoost in terms of both training time & accuracy (Ke et al., 2017). The algorithm uses a relatively recent technique of Gradient-based One-Side Sampling (GOSS) in order to filter out the instances of data for finding a split value unlike XGBoost which uses histogram based & pre-sorted algorithms in order to compute the ideal split (Ke et al., 2017). CatBoost is a recent open-source ML algorithm from Yandex. It can easily be integrated with modern deep learning frameworks such as Google’s TensorFlow or Apple’s Core ML. It performs well with diverse data types to help solve a wide range of problems that businesses face today, making an ideal candidate for this analysis. CatBoost is also appreciated for its ability to outperform all the algorithms when it comes to handling categorical features (Swalin, 2019).

**VotingClassifier**

The VotingClassifier ensemble method has shown promising results across various domains. In breast cancer prediction, it achieved 96% accuracy by combining ensemble algorithms (Rajani Uppara et al., 2023). For intrusion detection systems, a voting ensemble of decision trees, logistic regression, and support vector machines effectively classified the NSL-KDD dataset (Karuna G. Bagde & Atul D. Raut, 2024). In sentiment analysis of movie reviews, a voting approach using Naïve Bayes, K-nearest neighbor, random forest, and decision tree classifiers achieved 80.13% accuracy with normalized data (Achin Jain & Vanita Jain, 2021). For code smell detection, a heterogeneous voting ensemble of five base models (Decision Trees, Logistic Regression, Support Vector Machines, Multi-Layer Perceptron, and Stochastic Gradient Descent) outperformed individual models in detecting various code smells (Hamoud Aljamaan, 2021). These studies demonstrate the versatility and effectiveness of the VotingClassifier in improving predictive performance across different applications.
